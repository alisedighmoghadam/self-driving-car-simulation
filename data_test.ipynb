```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Load CSV file
df = pd.read_csv("./data/driving_log.csv")

# Select 4th column (steering angle in driving log)
col4 = df.iloc[:, 3]

# Plot histogram of column values
plt.hist(col4, bins=10,range=(-.25,.25)) 
plt.xlabel("Index")
plt.ylabel("Column 4 values")
plt.title("Column 4 Plot")
plt.show()
```
